```python
from sklearn.cluster import DBSCAN
import pandas as pd
from sklearn.preprocessing import MinMaxScaler


numerical_features = df.select_dtypes(include=['number']).columns
X = df[numerical_features]


scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN(eps=2, min_samples=5)
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
c1=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
dbscan.core_sample_indices_
```

```python
c1=pd.concat([df,c1],axis=1)
```

```python
fig, c1 = plt.subplots(figsize=(10, 6))

fig = plt.figure()
fig.set_size_inches(10, 6)
c1 = fig.add_subplot(111)
c1.scatter(df["Fresh"],df["Milk"],df["Grocery"],c=df["cluster"],cmap="tab10",alpha=0.7)
c1.set_xlabel("Fresh")
c1.set_ylabel("Milk")
c1.set_zlabel("Grocery")
c1.set_title("DBSCAN Clustering(3D View)")
plt.show()
```
